# Решаем систему  уравнений

```python
def equations(p):
    x, y = p
    f0 = 4*x**3 + 4*x*y + 20*x + 6*y - 6
    f1 = 2*x**2 + 6*x + 20*y
    return (f0,f1)

mx, my = fsolve(equations, (0,0))
print (mx,my)

```

# Вычисляем частные производные второго порядка

```python
# прозводные первого порядка
dfx = diff(x**4 + 2*(x**2)*y + 10*(x**2) + 6*x*y + 10*(y**2) - 6*x + 4, x)
dfy = diff(x**4 + 2*(x**2)*y + 10*(x**2) + 6*x*y + 10*(y**2) - 6*x + 4, y)
# производные второго порядка
dfx2 = diff(dfx, x)
dfy2 = diff(dfy, y)
dfxy = diff(dfx, y)
```

```python
print(dfx2)
print(dfy2)
print(dfxy)
```

# Проверяем необходимые условия

```python
#сначала найдем значения проивзодных второго порядка в найденной точке
A = 12*mx**2 + 4*my + 20
C = 20
B = 4*mx + 6

#выведем результат
print(A,B,C)

#чтобы в этой точке был экстремум должно выполняться условие :  AC – B**2 > 0
print(A*C - B **2 )

##условие выполняется и А больше нуля. Следовательно это Искомый минимум функции
```

# График

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

mx = np.arange(0, 1, 0.1)
my = np.arange(-1, 0, 0.1)

X, Y = np.meshgrid(mx, my)
func = X**4 + 2*(X**2)*Y + 10*(X**2) + 6*X*Y + 10*(Y**2) - 6*X + 4

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(func, X, Y)
plt.xlabel('function')
plt.ylabel('x')
plt.show()
```
